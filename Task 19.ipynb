## Q3. Missing & Duplicate Values Analysis

```python
# Checking missing values

print("Missing Values:\n")

print(df.isnull().sum())

#===============================================================

# Calculating missing value percentage

print("\nMissing Value Percentage:\n")

missing_percentage = (df.isnull().sum() / len(df)) * 100

print(missing_percentage)

#===============================================================

# Checking duplicate rows

duplicates = df.duplicated().sum()

print("\nDuplicate Rows :", duplicates)

#===============================================================

# Removing duplicate rows

if duplicates > 0:

    df = df.drop_duplicates()

    print("\nDuplicate rows removed successfully.")

else:

    print("\nNo duplicate rows found.")

#===============================================================

# Displaying new shape

print("\nNew Shape of Dataset:")

print(df.shape)
```

### Observations

- Checked the dataset for missing values in every column.
- Calculated the percentage of missing values.
- Checked whether duplicate rows were present.
- Removed duplicate rows (if any).
- The cleaned dataset will be used for further analysis.

## Q7. Correlation Heatmap

## Q9. Encoding Categorical Variables

```python
# Displaying categorical columns

categorical_columns = X.select_dtypes(include=["object", "string"]).columns

print("Categorical Columns:\n")

print(categorical_columns)

#===============================================================

# Displaying first two categorical columns before encoding

print("\nBefore Encoding:\n")

display(X[["fuel", "transmission"]].head())

#===============================================================

# Performing One-Hot Encoding

X_encoded = pd.get_dummies(X)

#===============================================================

# Displaying encoded columns

print("\nAfter Encoding:\n")

display(X_encoded.head())
```

### Observations

Before encoding, categorical columns contained text values.

One-Hot Encoding converted these text values into numerical columns containing 0s and 1s.

This makes the dataset suitable for machine learning algorithms.

## Q10. Feature Scaling

```python
# Creating scaler object

scaler = StandardScaler()

#===============================================================

# Selecting numeric independent columns

numeric_columns = X_encoded.select_dtypes(include=np.number).columns

#===============================================================

# Scaling numeric features

X_scaled = scaler.fit_transform(X_encoded[numeric_columns])

#===============================================================

# Converting into DataFrame

X_scaled = pd.DataFrame(X_scaled,
                        columns=numeric_columns)

#===============================================================

# Displaying first 5 rows

display(X_scaled.head())
```

### Observations

Feature Scaling converts numerical values into a common scale.

It improves the performance of many Machine Learning algorithms by ensuring that features with larger values do not dominate those with smaller values.

After scaling, the data has approximately:
- Mean = 0
- Standard Deviation = 1
