# SESSION 20 - LINEAR REGRESSION USING FORD CAR DATASET

```python
# Importing required libraries

import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score

import joblib
```

## Q1. Preparing Features

```python
# Separating Independent and Dependent Features

# Loading the dataset

df = pd.read_csv("ford_car_dataset.csv")

#===============================================================

X = df.drop("price", axis=1)

y = df["price"]
#===============================================================

print("Shape of X :", X.shape)

print("Shape of y :", y.shape)
```

### Observation

- Independent features contain all information required to predict the car price.
- The dependent feature is **price** because it is the value we want to predict.

## Q2. One-Hot Encoding

```python
# Finding categorical columns

categorical_columns = X.select_dtypes(include=["object", "string"]).columns

print("Categorical Columns:\n")

print(categorical_columns)
```

# Displaying before encoding

display(X.head())

```python
# Performing One-Hot Encoding

X = pd.get_dummies(X)

#===============================================================

# Converting boolean columns into integer

X = X.astype(int)

#===============================================================

# Displaying first 5 rows

display(X.head())
```

### Observation

The categorical columns have been converted into numerical columns using One-Hot Encoding. Machine Learning models require numerical inputs, so this step is necessary.

## Q3. Feature Scaling

```python
# Creating StandardScaler object

scaler = StandardScaler()

#===============================================================

# Selecting numerical columns

numerical_columns = ["year", "mileage", "tax", "mpg", "engineSize"]

#===============================================================

# Applying Standard Scaling

X[numerical_columns] = scaler.fit_transform(X[numerical_columns])

#===============================================================

# Displaying first 5 rows

display(X.head())
```

### Observation

Feature Scaling standardizes the numerical features so that they all have a similar scale.

This helps Linear Regression and many other Machine Learning algorithms perform better because no single feature dominates due to having larger values.

## Q4. Train-Test Split

```python
# Splitting dataset into training and testing data

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.33,
    random_state=42
)
```

```python
# Displaying shapes

print("X_train Shape :", X_train.shape)

print("X_test Shape :", X_test.shape)

print("y_train Shape :", y_train.shape)

print("y_test Shape :", y_test.shape)
```

### Observation

The dataset has been divided into two parts:

- **Training Data (67%)** → Used to train the Linear Regression model.
- **Testing Data (33%)** → Used to evaluate the model on unseen data.

Using separate training and testing data helps us measure how well the model generalizes to new data.

## Q5. Building Linear Regression Model

```python
# Creating Linear Regression model

model = LinearRegression()

#===============================================================

# Training the model

model.fit(X_train, y_train)
```

```python
# Displaying model intercept

print("Intercept:\n")

print(model.intercept_)

#===============================================================

# Displaying model coefficients

coefficients = pd.DataFrame({
    "Feature": X_train.columns,
    "Coefficient": model.coef_
})

print("\nCoefficients:\n")

display(coefficients)
```

### Observation

- The Linear Regression model has been trained successfully.
- The intercept is the predicted price when all feature values are zero.
- The coefficients indicate how much each feature affects the car price.
- Positive coefficients increase the predicted price, while negative coefficients decrease it.

## Q6. Making Predictions

```python
# Making predictions on test data

y_pred = model.predict(X_test)
```

```python
# Comparing actual and predicted values

comparison = pd.DataFrame({
    "Actual Price": y_test.values,
    "Predicted Price": y_pred
})

display(comparison.head(10))
```

## Q7. Model Evaluation using R² Score

```python
# Calculating R² Score

r2 = r2_score(y_test, y_pred)

print("R² Score :", r2)
```

### Observation

- R² Score measures how well the model predicts the target variable.
- A value closer to **1** indicates better performance.
- A value closer to **0** indicates poor performance.

Interpretation:

- If the R² score is above **0.80**, the model performs very well.
- If the score is between **0.60 and 0.80**, the model performs reasonably well.
- If the score is below **0.60**, the model may need better features or a different algorithm.

```python
## Q8. Saving the Model
```

```python
# Saving the trained Linear Regression model

joblib.dump(model, "LR_ford_car.pkl")

#===============================================================

# Saving the StandardScaler

joblib.dump(scaler, "scaler.pkl")

#===============================================================

# Saving column names

joblib.dump(X.columns.tolist(), "columns.pkl")

#===============================================================

print("Model, Scaler and Column Names saved successfully.")
```

### Observation

Three files have been saved:

- **LR_ford_car.pkl** → Trained Linear Regression model
- **scaler.pkl** → StandardScaler object
- **columns.pkl** → List of feature names

These files can be loaded later to make predictions without retraining the model.
