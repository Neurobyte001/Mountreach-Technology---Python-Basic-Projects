## Q2. Missing & Duplicate Values

- Find the total number of missing values.
- Check for duplicate rows.
- Remove duplicate rows if found.

```python
# Finding missing values in each column

print("Missing Values:\n")
display(df.isnull().sum())

#===============================================================

# Counting duplicate rows

dupe = df.duplicated().sum()

print(f"\nDuplicate Rows : {dupe}")

#===============================================================

# Removing duplicate rows if found

if dupe > 0:

    df = df.drop_duplicates()

    print("\nDuplicate rows removed successfully.")

else:

    print("\nNo duplicate rows found.")

#===============================================================

# Displaying new shape

print("\nNew Shape of Dataset:")

print(df.shape)
```

### Observations

- Checked all columns for missing values.
- Checked for duplicate rows.
- Duplicate rows were removed (if any).
- The dataset is now cleaner and ready for analysis.

## Q6. Correlation Heatmap

Create a heatmap of the correlation matrix for all numerical columns.
Identify which features are highly correlated with the target variable (price).

```python
# Selecting only numerical columns

numeric_df = df.select_dtypes(include=np.number)

#===============================================================

# Creating correlation matrix

correlation = numeric_df.corr()

#===============================================================

# Plotting heatmap

plt.figure(figsize=(8,6))

sns.heatmap(correlation,
            annot=True,
            cmap="coolwarm",
            fmt=".2f")

plt.title("Correlation Heatmap")

plt.show()

#===============================================================

# Correlation with Price

print("Correlation with Price:\n")

print(correlation["price"].sort_values(ascending=False))
```

### Observations

- The heatmap shows the relationship between all numerical features.
- Values close to **1** indicate a strong positive correlation.
- Values close to **-1** indicate a strong negative correlation.
- Values close to **0** indicate little or no relationship.
- The features having the highest correlation with **price** are the most useful for prediction.
