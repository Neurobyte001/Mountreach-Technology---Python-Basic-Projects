# SESSION 22 - HEART DISEASE CLASSIFICATION USING LOGISTIC REGRESSION

## Q1. Data Loading & Preprocessing


```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report
)

import joblib
```

## Q1. Load the Heart Disease Dataset and Preprocess


```python
# Loading the dataset

df = pd.read_csv("heart.csv")

#===============================================================

# Displaying first 5 rows

display(df.head())

#===============================================================

# Separating Independent and Dependent Features

X = df.drop("HeartDisease", axis=1)

y = df["HeartDisease"]

#===============================================================

# Displaying shapes

print("Shape of X :", X.shape)
print("Shape of y :", y.shape)

#===============================================================

# Finding categorical columns

categorical_columns = X.select_dtypes(include=["object", "string"]).columns

print("\nCategorical Columns:\n")
print(categorical_columns)

#===============================================================

# Performing One-Hot Encoding

X = pd.get_dummies(X, columns=categorical_columns)

# Converting Boolean values into Integer

X = X.astype(int)

#===============================================================

print("\nEncoded Dataset:\n")
display(X.head())
```

### Observations

* The dataset was loaded successfully.
* `HeartDisease` is the target variable.
* Categorical columns were converted into numerical format using One-Hot Encoding.
* The dataset is now ready for model training.


## Q2. Train-Test Split


```python
# Creating scaler object

scaler = StandardScaler()

#===============================================================

# Applying feature scaling

X_scaled = scaler.fit_transform(X)

#===============================================================

# Splitting the dataset

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.20,
    random_state=42
)

#===============================================================

# Displaying shapes

print("X_train Shape :", X_train.shape)
print("X_test Shape  :", X_test.shape)
print("y_train Shape :", y_train.shape)
print("y_test Shape  :", y_test.shape)
```

### Observations

* 80% of the data is used for training.
* 20% of the data is used for testing.
* Feature scaling was applied before splitting the dataset.


## Q3. Building Logistic Regression Model


```python
# Creating Logistic Regression model

model = LogisticRegression(max_iter=1000)

#===============================================================

# Training the model

model.fit(X_train, y_train)

print("Logistic Regression Model Trained Successfully!")
```

### Observations

* The Logistic Regression model was created successfully.
* The model learned patterns from the training data to classify patients as having heart disease (1) or not having heart disease (0).


## Q4. Making Predictions


```python
# Making predictions on test data

y_pred = model.predict(X_test)

#===============================================================

# Comparing actual and predicted values

comparison = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": y_pred
})

display(comparison.head(10))
```

### Observations

* The model generated predictions for unseen test data.
* Comparing actual and predicted values helps us understand where the model is correct or incorrect.


## Q5. Confusion Matrix


```python
# Creating confusion matrix

cm = confusion_matrix(y_test, y_pred)

#===============================================================

print("Confusion Matrix:\n")
print(cm)

#===============================================================

# Extracting values

TN = cm[0][0]
FP = cm[0][1]
FN = cm[1][0]
TP = cm[1][1]

print("\nTN (True Negative) :", TN)
print("FP (False Positive):", FP)
print("FN (False Negative):", FN)
print("TP (True Positive) :", TP)
```

### Confusion Matrix Explanation

* **TN**: Model correctly predicted No Heart Disease.
* **FP**: Model predicted Heart Disease, but the patient does not have it.
* **FN**: Model predicted No Heart Disease, but the patient actually has it.
* **TP**: Model correctly predicted Heart Disease.


## Q6. Model Evaluation Metrics


```python
# Calculating evaluation metrics

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

#===============================================================

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1 Score :", f1)

#===============================================================

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

### Metrics Explanation

* **Accuracy**: Overall percentage of correct predictions.
* **Precision**: How many predicted positive cases are actually positive.
* **Recall**: How many actual positive cases were identified.
* **F1 Score**: Balance between Precision and Recall.

A higher value indicates better model performance.


## Q7. Saving the Model


```python
# Saving the trained model

joblib.dump(model, "heart_model.pkl")

#===============================================================

# Saving the scaler

joblib.dump(scaler, "scaler.pkl")

#===============================================================

# Saving column names

joblib.dump(list(X.columns), "columns.pkl")

#===============================================================

print("Model, Scaler and Columns saved successfully!")
```

### Saved Files

* `heart_model.pkl` → Trained Logistic Regression model
* `scaler.pkl` → StandardScaler object
* `columns.pkl` → List of encoded feature columns


## Q8. Loading & Testing Saved Model


```python
# Loading saved objects

loaded_model = joblib.load("heart_model.pkl")
loaded_scaler = joblib.load("scaler.pkl")
loaded_columns = joblib.load("columns.pkl")

#===============================================================

# Creating sample patient record

sample_input = pd.DataFrame({
    "Age": [55],
    "Sex": ["M"],
    "ChestPainType": ["ASY"],
    "RestingBP": [140],
    "Cholesterol": [250],
    "FastingBS": [1],
    "RestingECG": ["Normal"],
    "MaxHR": [130],
    "ExerciseAngina": ["Y"],
    "Oldpeak": [2.3],
    "ST_Slope": ["Flat"]
})

#===============================================================

# Applying One-Hot Encoding

sample_input = pd.get_dummies(sample_input)

#===============================================================

# Aligning columns with training data

sample_input = sample_input.reindex(
    columns=loaded_columns,
    fill_value=0
)

#===============================================================

# Applying scaling

sample_input = loaded_scaler.transform(sample_input)

#===============================================================

# Making prediction

prediction = loaded_model.predict(sample_input)

#===============================================================

print("Prediction:", prediction[0])

if prediction[0] == 1:
    print("Heart Disease: YES")
else:
    print("Heart Disease: NO")
```

### Observations

* The saved model was loaded successfully.
* A sample patient record was preprocessed using the same encoding and scaling steps as the training data.
* The model predicted whether the patient is likely to have heart disease or not.


# SESSION 22 - STREAMLIT APPLICATION

## Q9 & Q10. Heart Disease Prediction Web App


```python
import streamlit as st
import pandas as pd
import joblib

#===============================================================

# Loading saved model and preprocessing objects

model = joblib.load("heart_model.pkl")
scaler = joblib.load("scaler.pkl")
columns = joblib.load("columns.pkl")

#===============================================================

# Page configuration

st.set_page_config(
    page_title="Heart Disease Predictor",
    layout="centered"
)

#===============================================================

# Title and description

st.title("❤️ Heart Disease Predictor")

st.write("Enter patient details below to predict heart disease.")

#===============================================================

# Input fields

age = st.number_input("Age", 1, 100, 55)

sex = st.selectbox("Sex", ["M", "F"])

chest_pain = st.selectbox(
    "Chest Pain Type",
    ["ATA", "NAP", "ASY", "TA"]
)

resting_bp = st.number_input("Resting Blood Pressure", 50, 250, 120)

cholesterol = st.number_input("Cholesterol", 0, 600, 200)

fasting_bs = st.selectbox("Fasting Blood Sugar > 120 mg/dl", [0, 1])

resting_ecg = st.selectbox(
    "Resting ECG",
    ["Normal", "ST", "LVH"]
)

max_hr = st.number_input("Maximum Heart Rate", 50, 250, 150)

exercise_angina = st.selectbox("Exercise Induced Angina", ["Y", "N"])

oldpeak = st.number_input("Oldpeak", -5.0, 10.0, 1.0)

st_slope = st.selectbox("ST Slope", ["Up", "Flat", "Down"])

#===============================================================

# Predict button

if st.button("Predict"):

    try:

        # Creating input DataFrame

        input_df = pd.DataFrame({
            "Age": [age],
            "Sex": [sex],
            "ChestPainType": [chest_pain],
            "RestingBP": [resting_bp],
            "Cholesterol": [cholesterol],
            "FastingBS": [fasting_bs],
            "RestingECG": [resting_ecg],
            "MaxHR": [max_hr],
            "ExerciseAngina": [exercise_angina],
            "Oldpeak": [oldpeak],
            "ST_Slope": [st_slope]
        })

        #=======================================================

        # Encoding categorical columns

        input_df = pd.get_dummies(input_df)

        #=======================================================

        # Aligning columns

        input_df = input_df.reindex(
            columns=columns,
            fill_value=0
        )

        #=======================================================

        # Scaling features

        input_scaled = scaler.transform(input_df)

        #=======================================================

        # Making prediction

        result = model.predict(input_scaled)[0]

        #=======================================================

        if result == 1:
            st.error("⚠️ Heart Disease: YES")
        else:
            st.success("✅ Heart Disease: NO")

    except Exception as e:
        st.error(f"Error: {e}")
```
