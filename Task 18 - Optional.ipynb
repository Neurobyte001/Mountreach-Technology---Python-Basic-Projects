## Q2. Missing & Duplicate Values

- Find the total number of missing values.
- Check for duplicate rows.
- Remove duplicate rows if found.

```python
# Finding missing values in each column

print("Missing Values:\n")
display(df.isnull().sum())

#===============================================================

# Counting duplicate rows

dupe = df.duplicated().sum()

print(f"\nDuplicate Rows : {dupe}")

#===============================================================

# Removing duplicate rows if found

if dupe > 0:

    df = df.drop_duplicates()

    print("\nDuplicate rows removed successfully.")

else:

    print("\nNo duplicate rows found.")

#===============================================================

# Displaying new shape

print("\nNew Shape of Dataset:")

print(df.shape)
```

## Q6. Correlation Heatmap

Create a heatmap of the correlation matrix for all numerical columns.
Identify which features are highly correlated with the target variable.

```python
# Selecting only numerical columns

num = df.select_dtypes(include=np.number)

#===============================================================

# Creating correlation matrix

correlation = num.corr()

#===============================================================

# Plotting heatmap

plt.figure(figsize=(8,6))

sns.heatmap(correlation,
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()

#===============================================================

# Correlation with BMi

print("Correlation with BMI:\n")

print(correlation["bmi"].sort_values(ascending=False))
```
