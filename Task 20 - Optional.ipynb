```python
import pandas as pd

#===============================================================

# Loading the dataset

df = pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")

#===============================================================

# Creating Independent and Dependent Features

X = df.drop("selling_price", axis=1)

y = df["selling_price"]

#===============================================================

print("Shape of X :", X.shape)
print("Shape of y :", y.shape)
```

```python
#===============================================================

# Finding categorical columns

categorical_columns = X.select_dtypes(include=["object", "string"]).columns

print("Categorical Columns:\n")
print(categorical_columns)

#===============================================================

# Applying One-Hot Encoding

X = pd.get_dummies(X, columns=categorical_columns)

# Converting Boolean values into Integer

X = X.astype(int)

#===============================================================

print(X.head())
```

```python
from sklearn.preprocessing import StandardScaler

#===============================================================

# Creating scaler object

scaler = StandardScaler()

#===============================================================

# Numerical columns

numerical_columns = [
    "year",
    "km_driven"
]

#===============================================================

# Applying Standard Scaling

X[numerical_columns] = scaler.fit_transform(X[numerical_columns])

#===============================================================

print(X.head())
```

```python
from sklearn.model_selection import train_test_split

#===============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.33,
    random_state=42
)

#===============================================================

print("X_train :", X_train.shape)
print("X_test  :", X_test.shape)

print("y_train :", y_train.shape)
print("y_test  :", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

#===============================================================

model = LinearRegression()

#===============================================================

model.fit(X_train, y_train)

#===============================================================

print("Intercept :")
print(model.intercept_)

print("\nCoefficients :")
print(model.coef_)
```

```python
#===============================================================

y_pred = model.predict(X_test)

#===============================================================

comparison = pd.DataFrame({

    "Actual Price": y_test.values,

    "Predicted Price": y_pred

})

print(comparison.head(10))
```

```python
from sklearn.metrics import r2_score

#===============================================================

score = r2_score(y_test, y_pred)

print("R² Score :", score)
```

The R² Score indicates how well the model predicts the selling price.

An R² value closer to 1 means better prediction accuracy,
while a value closer to 0 indicates poor prediction performance.

```python
import joblib

#===============================================================

joblib.dump(model, "LR_CarDekho.pkl")

joblib.dump(scaler, "Scaler_CarDekho.pkl")

joblib.dump(list(X.columns), "Columns_CarDekho.pkl")

#===============================================================

print("Model Saved Successfully!")
```
